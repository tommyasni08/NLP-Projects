```python
# Data Pre-processing
import numpy as np
import pandas as pd
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'./data.csv',error_bad_lines=False)
```

```python
df.dropna(inplace=True)
```

```python
password_tuple =np.array(df)
```

```python
password_tuple
```

```python
import random
random.shuffle(password_tuple)
```

```python
x= [labels[0] for labels in password_tuple]
y= [labels[1] for labels in password_tuple]
```

```python
# TF-IDF
```

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer= TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X =vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
print(X[0])
```

```python
first_document_vector = X[0]
first_document_vector.T.todense()
```

```python
df_2 =pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
df_2.sort_values(by=['TF-IDF'],ascending=False)
```

```python
# Apply ML model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0,multi_class='multinomial')
clf.fit(X_train,y_train)
```

```python
dt = np.array(['%123acbd'])
pred = vectorizer.transform(dt)
clf.predict(pred)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
(unique, counts) = np.unique(y_pred, return_counts=True)
frequencies = np.asarray((unique, counts)).T
frequencies
```

```python
# Model Performance
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
